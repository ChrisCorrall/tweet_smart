```python
from twython import Twython as tw
from twython import TwythonError
```

```python
# App keys and secrets stored as environment variables
import os
APP_KEY = os.environ.get('TWITTER_APP_KEY')
APP_SECRET = os.environ.get('TWITTER_APP_SECRET')

Access_Token = os.environ.get('TWITTER_ACCESS_TOKEN')
Access_Secret = os.environ.get('TWITTER_ACCESS_SECRET')

twitter = tw(APP_KEY, APP_SECRET, Access_Token, Access_Secret)
```

```python
# Pull the Twitter list 
my_user_id = 705469309
my_list_name = 'tweet_smart_list' #list name - create this on Twitter
my_list_slug = 'tweet-smart-list'
```

```python
# Get the list 
list_members = twitter.get_list_members(slug = my_list_slug, owner_id = my_user_id)
```

```python
import pandas as pd
```

```python
# Dataframe that List
df_list = pd.DataFrame.from_dict(list_members['users'], dtype=None, columns=None)
df_list.head()
```

```python
# Get rid of all the fluff - just get the useful numbers
mini_df = df_list.loc[:, ['screen_name', 'favourites_count', 'followers_count', 'statuses_count']]
mini_df.set_index('screen_name', True, False, True)
mini_df.head()
```
